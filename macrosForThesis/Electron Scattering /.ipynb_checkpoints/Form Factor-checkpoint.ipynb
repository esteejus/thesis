```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.integrate import quad
import sympy
#mp.dps = 50
```

```python
hc = 197.327053
pi = np.pi
r0 = 6.0
a  = 1.0
p0 = .001
rmax =  15.0

def den (r):
    if r < rmax:
        return p0/(1.0 + np.exp( (r - r0)/a) )
    else:
        return 0.0
    
rho = np.vectorize(den)
```

```python
x = np.linspace(0.0,15.0,1000)
y = rho(x)

plt.plot(x,y, '-' )
```
